## 3. Compare the Unhappiest and happiest countries in each region

```python
# Using dataset 2015 for this analysis

regions = data_2015['Region'].value_counts()  #finding all the regions in the dataset
print(regions)
```

```python
for i in regions.index:
    df_country = data_2015[data_2015['Region'] == i]
    df = df_country.reset_index(drop = True)
    minimum = df.min(axis = None)
    maximum = df.max(axis = None)
    minimum_happiness_score = minimum['Happiness Score']
    maximum_happiness_score = maximum['Happiness Score']
    
    print("Region :", i)
    print()
    
    df_min = df[df['Happiness Score'] == minimum_happiness_score]

    df_min = df_min[['Country', 'Happiness Rank', 'Happiness Score']]     #Unhappiest Country
    print('Unhappiest Country')
    print()
    print(df_min.to_string(index = False))
    print()
                                                                                 #vs                                                           

    df_max = df[df['Happiness Score'] == maximum_happiness_score]
    
    df_max = df_max[['Country', 'Happiness Rank', 'Happiness Score']]     #Happiest country
    print('Happiest Country')
    print()
    print(df_max.to_string(index = False))
    
    print()
    print()



    
```

## 4. Countries that became unhappy between 2015 to 2017

```python
#Year 2015
df_country_2015 = data_2015['Country']

df_happiness_score_2015 = data_2015['Happiness Score']

#Year 2017
df_country_2017 = data_2017['Country']

df_happiness_score_2017 = data_2017['Happiness.Score']

#Creating a dictionary for deleting the countries which are present in 2015 dataset but not present in 2017, or vice versa.
d = {}

for i in df_country_2015:
        d[i] = d.get(i,0) + 1
for i in df_country_2017:
        d[i] = d.get(i,0) + 1        

temp = []   #temp list for storing countries whose count is 1.
for i in d:
    if d[i] == 1:
        temp.append(i)

for i in temp:    #deleting those countries.
    del d[i]
```

```python
# printing the desrired comparision

countries = []
for x in d:
    df_15 = data_2015[data_2015['Country'] == x]
    i = df_15['Happiness Score']
    
    df_17 = data_2017[data_2017['Country'] == x]
    j = df_17['Happiness.Score']
    
    if float(i) > float(j):
        print("Country :", x, '   Happiness Score in 2015 :', format(float(i), '.3f'), '   Happiness Score in 2017 :', format(float(j), '.3f'))    
```

## 5. Find the country whose happiness decreased by the most amount

```python
max = 0   #variable 1
country = ''   #variable 2

for x in d:   #using previously created dictionary as comparing between 2015 with 2017.
    df_15 = data_2015[data_2015['Country'] == x]
    i = df_15['Happiness Score']
    
    df_17 = data_2017[data_2017['Country'] == x]
    j = df_17['Happiness.Score']
    
    temp = float(i) - float(j)

    if temp > max:
        
        country = x     #comparing and updating
        max = temp

    
print("Country :", country, "   The amount of happiness score is decreased by :", format(max, '.3f') )
```
